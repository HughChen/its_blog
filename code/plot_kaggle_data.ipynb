```python
import csv

models       = []
total_counts = []
model_counts = []
percentage   = []

with open('kaggle2017methods.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        models.append(row[0])
        total_counts.append(int(row[1]))
        model_counts.append(int(row[2]))
        percentage.append(float(row[3]))
models[0] = "Logistic Regression"
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcdefaults()
plt.style.use("ggplot")
fig, ax = plt.subplots()
fig.set_size_inches(8,4.5)

# Example data
y_pos = np.arange(len(models))
bars = ax.barh(y_pos, percentage, align='center', color="#66c2a5")

# Add percentages on the bars
i = 0
for bar in bars:
    height = bar.get_height(); width = bar.get_width()
    plt.text(width-1.5, 
             bar.get_y() + height/2.0, 
             "{}%".format("%.2f" % percentage[i]), 
             ha='right', va='center', color="white")
    i += 1

ax.set_yticks(y_pos)
ax.set_yticklabels(models)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Percentage')
ax.set_title('Data Science Methods (Kaggle 2017)')

plt.tight_layout()
plt.savefig("images/methods.png", transparent=True)

plt.show()
```
